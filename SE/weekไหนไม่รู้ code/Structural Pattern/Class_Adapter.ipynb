```python
# Target
class Printer:
    def print_message(self, message):
        print(f"Printer: {message}")

# Client code that uses the Printer interface
def main():
    # Using the adapted Printer interface
    printer = Printer()
    printer.print_message("Hello, Adapter Pattern!")

if __name__ == "__main__":
    main()
```

```python
# Adaptee
class TempPrinter:
    def specific_print_message(self, message):
        print(f"Temporary Printer: {message}")

# Class Adapter lets the TempPrinter respond to requests on the Printer
class PrinterAdapter(Printer, TempPrinter):
    def print_message(self, message):
        # Using the existing method from LegacyPrinter
        self.specific_print_message(message)

# Client code that uses the Printer interface
def main():
    # Using the adapted Printer interface
    printer = PrinterAdapter()
    printer.print_message("Hello, Adapter Pattern!")

if __name__ == "__main__":
    main()
```
